```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,10)
y = np.sin(x)
z = np.cos(x)

plt.plot(x,y,'b',x,z,'r')
plt.xlabel('Radians');
plt.ylabel('Value');
plt.title('Plotting Demonstration')
plt.legend(['Sin','Cos'])
plt.grid()
```

```python
emas1 = 25
hargaBeliEmas1 = 650000
hargaJualEmas1 = 685000

selisihHarga1 = hargaJualEmas1 - hargaBeliEmas1
untungRupiah1 = emas1 * selisihHarga1
untungPersen1 = (selisihHarga1 / hargaBeliEmas1) * 100

print("Kasus 1")
print(f"Keuntungan dalam Rupiah = Rp. {untungRupiah1}")
print(f"Keuntungan dalam Persen = {untungPersen1:.2f}%")


emas2 = 15
hargaBeliEmas2 = 685000
hargaJualEmas2 = 715000

totalEmas = emas1 + emas2
totalModal = (emas1 * hargaBeliEmas1) + (emas2 * hargaBeliEmas2)
totalKeuntungan = (totalEmas * hargaJualEmas2) - totalModal
untungPersen2 = (totalKeuntungan / totalModal) * 100

print("\nKasus 2")
print(f"Keuntungan dalam Rupiah = Rp. {totalKeuntungan}")
print(f"Keuntungan dalam Persen = {untungPersen2:.2f}%")
```
